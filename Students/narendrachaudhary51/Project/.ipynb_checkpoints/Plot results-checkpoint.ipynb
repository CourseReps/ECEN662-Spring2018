```python
import numpy as np
from scipy.stats import powerlaw, norm, uniform, poisson
import matplotlib.pyplot as plt
import pandas as pd
%matplotlib inline  
```

```python
df = pd.DataFrame.from_csv("Test_results.csv")
x = df.loc[(df['a0'] == 10) & (df['a1'] == 1) & (df['Mean difference'] == 0.5)]['K'].as_matrix()

a0 = 40
a1 = 1
y1 = df.loc[(df['a0'] == a0) & (df['a1'] == a1) & (df['Mean difference'] == 1)]['Relative % Error'].as_matrix()
y2 = df.loc[(df['a0'] == a0) & (df['a1'] == a1) & (df['Mean difference'] == 2)]['Relative % Error'].as_matrix()
y3 = df.loc[(df['a0'] == a0) & (df['a1'] == a1) & (df['Mean difference'] == 3)]['Relative % Error'].as_matrix()
y4 = df.loc[(df['a0'] == a0) & (df['a1'] == a1) & (df['Mean difference'] == 4)]['Relative % Error'].as_matrix()
y5 = df.loc[(df['a0'] == a0) & (df['a1'] == a1) & (df['Mean difference'] == 5)]['Relative % Error'].as_matrix()
y6 = df.loc[(df['a0'] == a0) & (df['a1'] == a1) & (df['Mean difference'] == 6)]['Relative % Error'].as_matrix()
y7 = df.loc[(df['a0'] == a0) & (df['a1'] == a1) & (df['Mean difference'] == 7)]['Relative % Error'].as_matrix()
#y8 = df.loc[(df['a0'] == a0) & (df['a1'] == a1) & (df['Mean difference'] == 8)]['Relative % Error'].as_matrix()

plt.plot(x,y1, marker = 'o', label = 'Mean diff = 1')
plt.plot(x,y2, marker = 'o', label = 'Mean diff = 2')
plt.plot(x,y3, marker = 'o', label = 'Mean diff = 3')
plt.plot(x,y4, marker = 'o', label = 'Mean diff = 4')
plt.plot(x,y5, marker = 'o', label = 'Mean diff = 5')
plt.plot(x,y6, marker = 'o', label = 'Mean diff = 6')
plt.plot(x,y7, marker = 'o', label = 'Mean diff = 7')
#plt.plot(x,y8, marker = 'o', label = 'Mean diff = 8')

plt.title('Relative Error vs sum of K powerlaws with a0 = {}, a1 = {}'.format(a0,a1))
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.xlabel('Sum of K variables')
plt.ylabel('Relative % Error')
plt.show()
```
