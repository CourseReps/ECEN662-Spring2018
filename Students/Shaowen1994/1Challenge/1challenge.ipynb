# Challenge 1: Binary Detection 1

The first step is to import relevant libraries.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

The second step is to import data from the given file.

```python
df = pd.DataFrame.from_csv("1challenge.csv")
df0 = df.loc[df['label'] == 1.0]
df1 = df.loc[df['label'] == 0.0]
dftest = df.loc[~((df['label'] == 0.0) | (df['label'] == 1.0))]
print(df0.shape)
print(df1.shape)
print(dftest.shape)
```

One can use the data in their original Pandas DataFrame format, or one can transform these objects into Numpy-arrays.

```python
TrainingData0 = df0.as_matrix(columns=None)
TrainingData1 = df1.as_matrix(columns=None)
TestData = dftest.as_matrix(columns=['Y0', 'Y1'])
```

It may be helpful to visualize the data.

```python
plt.plot(TrainingData0[:,0], TrainingData0[:,1], 'x', color='r')
plt.plot(TrainingData1[:,0], TrainingData1[:,1], 'x', color='b')
plt.plot(TestData[:,0], TestData[:,1], 'o', color='k')
plt.axis('equal')
plt.show()
```

After creating an algorithm and generating labels, one should update the original CSV file.

```python
def gaussian(data):        # Gaussian fitting
    mean = np.mean(data)
    var = np.var(data)
    x = np.arange(min(data),max(data),0.01)
    y = np.exp(-(x-mean)**2/(2*var))/(np.sqrt(2*np.pi*var))
    return x,y
    
plt.subplot(221)
plt.hist(TrainingData0[:,0],200,normed=1,color='b',alpha=.5)
x,y = gaussian(TrainingData0[:,0])
plt.plot(x,y,color='k')
plt.subplot(222)
plt.hist(TrainingData0[:,1],200,normed=1,color='g',alpha=.5)
x,y = gaussian(TrainingData0[:,1])
plt.plot(x,y,color='k')
plt.subplot(223)
plt.hist(TrainingData0[:,0],200,normed=1,color='r',alpha=.5)
x,y = gaussian(TrainingData1[:,0])
plt.plot(x,y,color='k')
plt.subplot(224)
plt.hist(TrainingData0[:,0],200,normed=1,color='y',alpha=.5)
x,y = gaussian(TrainingData1[:,1])
plt.plot(x,y,color='k')

plt.show()
```

From above we can see that the training datas fit well with the gaussian fitting, so we can assume that the training datas with the same label follow a two-dimensional Gaussian distribution. The ratio of the number of datas with label "0" to that of datas with label "1" is 2/3, so we can assume that the ratio of posterior probabilities of label "1" and "0" is 2/3. 

```python
pos_ratio = 2/3
label = np.ones(len(TestData[:,0]))

def gaussian_2d(x,y,data_1,data_2):
    mean_1 = np.mean(data_1)
    mean_2 = np.mean(data_2)
    var_1 = np.var(data_1)
    var_2 = np.var(data_2)
    cov = np.cov(data_1,data_2)
    p = cov[1,0]/np.sqrt(var_1*var_2)
    prob = np.exp(-1/(2*(1-p**2))*((x-mean_1)**2/var_1+(y-mean_2)**2/var_2-2*p*(x-mean_1)*(y-mean_2)/np.sqrt(var_1*var_2)))/(2*np.pi*np.sqrt((1-p**2)*var_1*var_2))
    return prob

def distribution_0(x,y):   #probability distribution of patern 0
    prob = gaussian_2d(x,y,TrainingData0[:,0],TrainingData0[:,1])    
    return prob
def distribution_1(x,y):   #probability distribution of patern 1
    prob = gaussian_2d(x,y,TrainingData1[:,0],TrainingData1[:,1])    
    return prob

for i in range (0,len(TestData[:,0])):
    if (distribution_1(TestData[i,0],TestData[i,1])/distribution_0(TestData[i,0],TestData[i,1]) >= pos_ratio):  #Bayes Classifier
        label[i] = 1
    else:
        label[i] = 0 

TestData_1  = np.c_[TestData,label]
  
for i in range (0,len(TestData[:,0])):
    if TestData_1[i,2] == 0:
        plt.plot(TestData_1[i,0], TestData_1[i,1], 'x', color='r')
    else:
        plt.plot(TestData_1[i,0], TestData_1[i,1], 'o', color='b',alpha = 0.5)
plt.axis('equal')
plt.show()
```

```python
dftest_1 = pd.DataFrame(TestData_1,columns = ['Y0','Y1','label'] )
```

```python
df = pd.concat([df0, df1, dftest_1], join='outer', ignore_index=True)
df.to_csv("1challenge.csv")
```
