```python
import pandas as pd
df = pd.read_csv('2challenge.csv',index_col=0)
# na filling w average all columns except the label column
df.iloc[:,:8] = df.fillna(df.mean()).iloc[:,:8]
df
```

```python
# Dividing Data into Training and Test Sets
TrainingData = df.iloc[:10000,:8]
TrainingLabels = df.iloc[:10000,8]
TestData = df.iloc[10000:,:8]
```

```python
# 10 fold CV to overcome overfitting
from sklearn.model_selection import KFold
kfold = KFold(n_splits=10)
```

```python
# Logistic Regression
from sklearn.linear_model import LogisticRegression
from sklearn import model_selection
model = LogisticRegression()
results = model_selection.cross_val_score(model, TrainingData, TrainingLabels, cv=kfold)
print("Accuracy: %.3f%% (%.3f%%)" % (results.mean()*100.0, results.std()*100.0))
```

```python
# Random Forests
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier(n_estimators=10)
results = model_selection.cross_val_score(model, TrainingData, TrainingLabels, cv=kfold)
print("Accuracy: %.3f%% (%.3f%%)" % (results.mean()*100.0, results.std()*100.0))
```

```python
# SVM Classifier
from sklearn import svm
model = svm.SVC()
results = model_selection.cross_val_score(model, TrainingData, TrainingLabels, cv=kfold)
print("Accuracy: %.3f%% (%.3f%%)" % (results.mean()*100.0, results.std()*100.0))
```

```python
# It seems SVM gives the best
# result, so I will choose that as the final
# learning algorithm
model = svm.SVC()
model.fit(TrainingData, TrainingLabels)
df.iloc[10000:,8] = model.predict(TestData)
```
