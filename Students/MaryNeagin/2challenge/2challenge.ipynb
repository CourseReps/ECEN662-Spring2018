```python
#!/usr/bin/env python3
# -*- coding: utf-8 -*-
"""
Created on Wed Feb 14 11:23:28 2018

@author: maryneagin
"""
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.DataFrame.from_csv('2challenge.csv')
df0 = df.loc[df['label'] == 1.0]
df1 = df.loc[df['label'] == 0.0]


dftest = df.loc[~((df['label'] == 0.0) | (df['label'] == 1.0))]
print(df0.shape)
print(df1.shape)
print(dftest.shape)

TrainingData0 = df0.as_matrix(columns=None)
TrainingData1 = df1.as_matrix(columns=None)
TestData = dftest.as_matrix(columns=['Y0', 'Y1','Y2','Y3','Y4','Y5','Y6','Y7'])



plt.plot(TrainingData0[:,0], TrainingData0[:,1],TrainingData0[:,2],TrainingData0[:,3],TrainingData0[:,4],TrainingData0[:,5],TrainingData0[:,6],TrainingData0[:,7],'x', color='r')
plt.plot(TrainingData1[:,0], TrainingData1[:,1],TrainingData1[:,2],TrainingData1[:,3],TrainingData1[:,4],TrainingData1[:,5],TrainingData1[:,6],TrainingData1[:,7], 'x', color='b')
plt.plot(TestData[:,0], TestData[:,1],TestData[:,2],TestData[:,3],TestData[:,4],TestData[:,5],TestData[:,6],TestData[:,7], 'o', color='k')


#label 0
m_label0 = np.mean(TrainingData0, axis = 0)[0:8]  # mean of label 0 data\n",
mean_label0= np.sum(m_label0[0:6])/6;
print(m_label0)
print(mean_label0)
covariance_matrix_label0 = np.cov([TrainingData0[:,0],TrainingData0[:,1],TrainingData0[:,2],TrainingData0[:,3],TrainingData0[:,4],TrainingData0[:,5],TrainingData0[:,6],TrainingData0[:,7]])    # covariance matrix of label 0 data\n",
variance0_label0=np.var(TrainingData0[:,0])
variance1_label0 = np.var(TrainingData0[:,1])  
variance2_label0= np.var(TrainingData0[:,2]) 
variance3_label0 = np.var(TrainingData0[:,3]) 
variance4_label0 = np.var(TrainingData0[:,4]) 
variance5_label0 = np.var(TrainingData0[:,5]) 
variance6_label0 = np.var(TrainingData0[:,6]) 
variance7_label0 = np.var(TrainingData0[:,7]) 
#covariance1 = np.cov(TrainingData1[:,0],TrainingData1[:,1])    # covariance matrix of label 1 data\n",
#print(covariance_matrix_label0)
print(variance0_label0)
print(variance1_label0)
print(variance2_label0)
print(variance3_label0)
print(variance4_label0)
print(variance5_label0)
print(variance6_label0)
print(variance7_label0)

#label 1
m_label1 = np.mean(TrainingData1, axis = 0)[0:8]  # mean of label 1 data\n",
mean_label1= np.sum(m_label1[0:6])/6;
print(m_label1)
print(mean_label1)
#covariance_matrix_label1 = np.cov([TrainingData1[:,0],TrainingData1[:,1],TrainingData1[:,2],TrainingData1[:,3],TrainingData1[:,4],TrainingData1[:,5],TrainingData1[:,6],TrainingData1[:,7]])    # covariance matrix of label 0 data\n",
variance0_label1=np.var(TrainingData1[:,0])
variance1_label1 = np.var(TrainingData1[:,1])  
variance2_label1= np.var(TrainingData1[:,2]) 
variance3_label1 = np.var(TrainingData1[:,3]) 
variance4_label1 = np.var(TrainingData1[:,4]) 
variance5_label1 = np.var(TrainingData1[:,5]) 
variance6_label1 = np.var(TrainingData1[:,6]) 
variance7_label1 = np.var(TrainingData1[:,7]) 

#covariance1 = np.cov(TrainingData1[:,0],TrainingData1[:,1])    # covariance matrix of label 1 data\n",
#print(covariance_matrix_label1)
print(variance0_label1)
print(variance1_label1)
print(variance2_label1)
print(variance3_label1)
print(variance4_label1)
print(variance5_label1)
print(variance6_label1)
print(variance7_label1)

#corrcoef_label0 = np.corrcoef([TrainingData0[:,0],TrainingData0[:,1],TrainingData0[:,2],TrainingData0[:,3],TrainingData0[:,4],TrainingData0[:,5],TrainingData0[:,6],TrainingData0[:,7]])
#print(corrcoef_label0)
mean=mean_label0+mean_label1
print(mean)


m=np.isnan(TestData)
m1=m.sum(axis=1)
print(m1)


p1=np.nonzero(m1)
print(p1)

p=np.arange(5000)
p2=np.delete(p,p1)
p1=np.delete(p,p2)
sum_label8=TestData[p2,:].sum(axis=1)/8
sum_label6=TestData[p1,0:6].sum(axis=1)/6
threshold=mean/2


sum_label=np.zeros(TestData.shape[0])
sum_label[p2]=sum_label8
sum_label[p1]=sum_label6

labels = np.zeros(TestData.shape[0])
for i in range(TestData.shape[0]):
   if sum_label[i] > threshold:     #decision rule
       labels[i]=1
   else:
            labels[i] = 0
print(labels)

predictions_ans=np.array(labels)
predictions_ans=predictions_ans.reshape(predictions_ans.shape[0],1)
TestData=np.append(TestData,predictions_ans,axis=1)
dfdata=pd.DataFrame(TestData,columns=['Y0', 'Y1','Y2','Y3','Y4','Y5','Y6','Y7','label'])
df = pd.concat([df0, df1, dfdata], join='outer', ignore_index=True)
df.to_csv('2challenge.csv')
```
